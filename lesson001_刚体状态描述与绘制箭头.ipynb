### Python练习

#### 点的建立与运算

##### 利用list类型

```python
# 建立连个点
Point_A=[1,2]
Point_B=[2,3]
Line_AB=[Point_B[0]-Point_A[0],Point_B[1]-Point_A[1]]
print(Line_AB)
```

##### 利用numpy类型

```python
#建立两个点
import numpy as np
Point_A=np.array([1,2])
Point_B=np.array([2,3])
Line_AB=Point_B-Point_A
print(Line_AB)
```

##### 错误方法

```python
import matplotlib.pyplot as plt
import numpy as np
ax = plt.axes()
print(Point_A)
print(Point_B)
ax.arrow(Point_A[0],Point_A[1],Point_B[0],Point_B[1])
plt.scatter(Point_B[0],Point_B[1])
plt.grid() # 增加网格
plt.show()
```

是坐标轴相等`plt.axis('equal')`

```python
import matplotlib.pyplot as plt
import numpy as np
ax = plt.axes()
print(Point_A)
print(Point_B)
ax.arrow(Point_A[0],Point_A[1],Point_B[0],Point_B[1])
plt.scatter(Point_B[0],Point_B[1])
plt.grid() # 增加网格
plt.axis('equal')
plt.show()
```

##### 正确方法

```python
import matplotlib.pyplot as plt
import numpy as np
ax = plt.axes()
print(Point_A)
print(Point_B)
# ax.arrow(Point_A[0],Point_A[1],Point_B[0],Point_B[1])
ax.arrow(Point_A[0],Point_A[1],Line_AB[0],Line_AB[1])
plt.scatter(Point_B[0],Point_B[1])
plt.grid()
plt.axis('equal')
plt.show()
```

##### 修改样式

```python
import matplotlib.pyplot as plt
import numpy as np
ax = plt.axes()
print(Point_A)
print(Point_B)
ax.arrow(Point_A[0],Point_A[1],Line_AB[0],Line_AB[1],width=0.01, head_width=0.05, head_length=0.1, fc='r', ec='b',length_includes_head=True)
plt.scatter(Point_A[0],Point_A[1],color='blue',s=30*2**2)
plt.grid()
plt.axis('equal')
plt.show()
```
